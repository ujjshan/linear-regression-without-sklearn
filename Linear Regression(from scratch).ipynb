```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# Reading data (sample data)
data= pd.read_csv('C:\\Users\\Ujjwal Shandilya\\datasets\\headbrain.csv')
print(data.shape)
data.head()
```

```python
# collecting X and Y for linear Regression
x = data['Head Size(cm^3)'].values
y = data['Brain Weight(grams)'].values
```

```python
# Mean of x and y
mean_x = np.mean(x)
mean_y = np.mean(y)

# Total no. of values
m = len(x)

#calculating slope(m/b1) and constant(c/b0) for y=mx+c
numerator = 0
denominator= 0
for i in range(m):
    numerator += (x[i]-mean_x) * (y[i]-mean_y)
    denominator +=(x[i]-mean_x)**2
b1 = numerator/denominator
b0 = mean_y - (b1 * mean_x)

print(b1,b0)
```

```python
# ploting values and Regression line

max_x = np.max(x) + 100
min_x = np.min(x) - 100

# collecting line values of x and y

X = np.linspace(min_x, max_x, 1000)
Y = b0 + b1*X

# ploting line
plt.plot(X,Y,color='blue', label='Regression line')
#ploting scatter points
plt.scatter(x,y,c='red', label='Scatter points')

plt.xlabel('Head size in cm^3')
plt.ylabel('Brain wieght in grams')
plt.legend()
plt.show()
```

```python
# goodness of the fitted line by r squared method
#ss_t: total sum of square
#ss_r: total sum of square of residual
#r2=1-(ss_r/ss_t)

ss_t = 0
ss_r = 0
for i in range(m):
    y_pred =  b0 + b1*x[i]
    ss_t += (y[i]-mean_y)**2
    ss_r += (y[i]-y_pred)**2
r2 = 1-(ss_r/ss_t)
print(r2)
```
